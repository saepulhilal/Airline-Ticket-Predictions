## Data Inference


Membuat 10 data dumy, data ini akan diprediksi harganya

## Import Libaries

```python
import pickle
import pandas as pd
import numpy as np
```

## Import Model

```python
with open("model.pkl", "rb") as f:
    model = pickle.load(f)
```

## Create Data

```python
np.random.seed(40)

airline = ['Air_India', 'Vistara', 'SpiceJet', 'Indigo', 'AirAsia', 'GO_FIRST']
source_city = ['Kolkata', 'Chennai', 'Bangalore', 'Hyderabad', 'Mumbai', 'Delhi']
departure_time = ['Evening', 'Early_Morning', 'Morning', 'Afternoon', 'Night', 'Late_Night']
stops = ['one', 'two_or_more', 'zero']
arrival_time = ['Afternoon', 'Night', 'Evening', 'Early_Morning', 'Morning', 'Late_Night']
destination_city = ['Delhi', 'Mumbai', 'Chennai', 'Bangalore', 'Kolkata', 'Hyderabad']
type = ['Economy', 'Business']


data = {
    'airline': np.random.choice(airline, 10),
    'source_city': np.random.choice(source_city, 10),
    'departure_time': np.random.choice(departure_time, 10),
    'stops': np.random.choice(stops, 10),
    'arrival_time': np.random.choice(arrival_time, 10),
    'destination_city': np.random.choice(destination_city, 10),
    'type': np.random.choice(type, 10),
    'duration': np.random.uniform(0.83, 30, 10),
    'days_left': np.random.uniform(1, 49, 10)
    }

data = pd.DataFrame(data)
```

## Menampilkan Data

```python
data
```

## Prediction

```python
# Hasil prediksi
predict = model.predict(data)
data['price_predict'] = predict
data
```
